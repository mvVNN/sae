```python
##### Debut requetes 

import pymongo
import sqlite3
import pandas as pd

client = pymongo.MongoClient()
db = client.SAE2

collection_clients = db.Clients
collection_employes = db.Employe

# Agr√©gation pour calculer le nombre de commandes par client
pipeline = [
    {
        "$project": {
            "customerNumber": 1,
            "nb_commandes": {"$size": "$Commande"}
        }
    },
    {
        "$match": {
            "nb_commandes": 0
        }
    }
]

# Ex√©cution de l'agr√©gation
q1 = collection_clients.aggregate(pipeline)

# Cr√©ation d'un DataFrame Pandas √† partir des r√©sultats
df = pd.DataFrame(list(q1))

# Affichage du DataFrame
print(df)
```

```python
# Op√©ration d'agr√©gation avec $lookup pour joindre Employe et Clients

# Op√©ration d'agr√©gation avec $lookup pour joindre Employe et Clients
pipeline2 = [
    {
        "$lookup": {
            "from": "Clients",
            "localField": "employeeNumber",
            "foreignField": "salesRepEmployeeNumber",
            "as": "clients"
        }
    },
    {
        "$unwind": "$clients"
    },
    {
        "$group": {
            "_id": {
                "employeeNumber": "$employeeNumber",
                "lastName": "$lastName",
                "firstName": "$firstName"
            },
            "nombre_clients": {"$sum": 1},
            "nombre_commandes": {"$sum": {"$size": "$clients.Commande"}},
            "montant_total_commandes": {"$sum": {"$sum": "$clients.Commande.priceEach"}}
        }
    },
    {
        "$project": {
            "_id": 0,
            "employeeNumber": "$_id.employeeNumber",
            "lastName": "$_id.lastName",
            "firstName": "$_id.firstName",
            "nombre_clients": 1,
            "nombre_commandes": 1,
        }
    }
]

# Ex√©cution de l'agr√©gation
cursor = collection_employes.aggregate(pipeline2)

# Cr√©ation d'un DataFrame Pandas √† partir des r√©sultats
df = pd.DataFrame(list(cursor))

# Affichage du DataFrame
print(df)
```
